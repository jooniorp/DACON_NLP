## 데이터 변형 없이 기본 모델 finetuning

```python
#질문, 답변 모든 경우의 수로 딕셔너리 객체로 리스트에 삽입
#모델의 입력으로 넣기 위한 형식으로 전처리
#원본 데이터 644개 1행당 10개 => 6440개 데이터 답변 모든 경우의 수로 딕셔너리 객체로 리스트에 삽입
#원본 데이터 644개 1행당 10개 => 6440개 데이터
#id를 같이 넣은 이유는 밑의 data augmentation할 때 중복된 질문을 방지하기 위함
from tqdm import tqdm

data = []

for _, row in tqdm(df.iterrows()):
    for q_col in ['질문_1', '질문_2']:
        for a_col in ['답변_1', '답변_2', '답변_3', '답변_4', '답변_5']:
            data.append({'question':row[q_col], 'answer':row[a_col], 'id':row['id']})

data[-5:]
```

```python
# data augmentation 두 가지의 질문을 접속사와 결합
import random

def random_concat(data):
    s1, s2 = random.sample(data, 2)

    conjunctions = ['와', '그리고', '또한','랑','및']
    conjunction = random.choice(conjunctions)

    if s1['id'] != s2['id']:#동일한 질문이 이어지는 것을 방지
        concat_q = s1['question'] + " " + conjunction + " " + s2['question']
        concat_a = "1."+s1['answer'] + " " + "2." + s2['answer']

    else:
        return 0

    aug_data = {'question':concat_q, 'answer':concat_a}
    return aug_data
```

```python
random_concat(data)
```

```python
#n개 만큼 data augmentation
n = 500
aug_datas = [result for _ in range(n) if (result := random_concat(data)) != 0]
aug_datas
```
